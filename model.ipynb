```python
from sklearn.model_selection import GroupShuffleSplit

groups = df["collection_id"]
gss = GroupShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in gss.split(df, groups=groups):
    train_df = df.iloc[train_index]
    test_df = df.iloc[test_index]
```

```python
xtrain= train_df.drop(columns="last_sale_price")
ytrain= train_df["last_sale_price"]
```

```python
xtest= test_df.drop(columns="last_sale_price")
ytest= test_df["last_sale_price"]
```

```python
# changing there data type for object to integer
xtrain['contract_type'] = xtrain['contract_type'].map({'own': 1, 'default': 0})
xtrain['verification_status'] = xtrain['verification_status'].map({'approved': 1, 'verified': 0})
xtest['contract_type'] = xtest['contract_type'].map({'own': 1, 'default': 0})
xtest['verification_status'] = xtest['verification_status'].map({'approved': 1, 'verified': 0})
```

```python
xtrain.info()
```

```python
xtrain['last_sale_date'] = pd.to_datetime(xtrain['last_sale_date'], format='%Y-%m')
xtest['last_sale_date'] = pd.to_datetime(xtest['last_sale_date'], format='%Y-%m')
```

```python
xtrain['creation_date'] = pd.to_datetime(xtrain['creation_date'], format='%Y-%m')
xtest['creation_date'] = pd.to_datetime(xtest['creation_date'], format='%Y-%m')
```

```python
xtrain.last_sale_date.dtype
```

```python
xtrain.drop(columns=['last_sale_date', 'creation_date'], inplace=True)
xtest.drop(columns=['last_sale_date', 'creation_date'], inplace=True)
```

```python
ytrain.head()
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import RandomizedSearchCV
import time

nn = MLPRegressor(random_state=7, max_iter = 200)

# specify the hyperparameters and their values
# 4 x 3 x 2 = 24 combinations in the grid
param_grid = {
    'hidden_layer_sizes': [(100,)],
    'activation': ['relu'],
    'solver': ['lbfgs', 'adam']
}

# we'll use 5-fold cross-validation
grid_search = RandomizedSearchCV(nn, param_grid, cv=5,
                           scoring='neg_root_mean_squared_error', 
                           return_train_score=True,
                          n_jobs = -1) 

start = time.time()
grid_search.fit(xtrain, ytrain)
nn_time = time.time() - start
print(f"Took {nn_time} seconds")
```

```python
nn_train_score = grid_search.best_score_
nn_train_score
```

```python
import xgboost as xgb

# convert data into DMatrix
dtrain = xgb.DMatrix(xtrain, label=ytrain, enable_categorical = 'True')
dtest = xgb.DMatrix(xtest, label=ytest, enable_categorical = 'True')

# define the parameters for XGBoost
params = {'objective': 'reg:squarederror', # regression objective
          'eval_metric': 'rmse', # evaluation metric
          'eta': 0.01, # learning rate
          'max_depth': 10, # maximum depth of the tree
          'subsample': 0.8, # subsample ratio
          'colsample_bytree': 0.8# column sample by tree ratio
         }

# train the model using the defined parameters
xg_reg = xgb.train(params, dtrain, num_boost_round=1000, enable_categorical = True)
```

```python
# predict on the test set
y_pred = xg_reg.predict(dtest)
```

```python
# calculate the root mean squared error
from sklearn.metrics import mean_squared_error
rmse = np.exp(-mean_squared_error(ytest, y_pred, squared=False)/10)
print("Root Mean Squared Error: ", rmse)
```

```python
# OpenRarity version 0.4.0-beta
from open_rarity import (
    Collection,
    Token,
    RarityRanker,
    TokenMetadata,
    StringAttribute,
)
from open_rarity.models.token_identifier import EVMContractTokenIdentifier
from open_rarity.models.token_standard import TokenStandard

# Create OpenRarity collection object and provide all metadata information
collection = Collection(
    name="My Collection Name",
    attributes_frequency_counts={
        "hat": {"cap": 1, "visor": 2},
        "shirt": {"blue": 2, "green": 1},
    },
    tokens=[
        Token(
            token_identifier=EVMContractTokenIdentifier(
                contract_address="0xa3049...", token_id=1
            ),
            token_standard=TokenStandard.ERC721,
            metadata=TokenMetadata(
                string_attributes={
                    "hat": StringAttribute(name="hat", value="cap"),
                    "shirt": StringAttribute(name="shirt", value="blue"),
                }
            ),
        ),
    ],
)  # Replace inputs with your collection-specific details here


# Generate scores for a collection
ranked_tokens = RarityRanker.rank_collection(collection=collection)

# Iterate over the ranked and sorted tokens
for token_rarity in ranked_tokens:
    token_id = token_rarity.token.token_identifier.token_id
    rank = token_rarity.rank
    score = token_rarity.score
    print(f"\tToken {token_id} has rank {rank} score: {score}")
```
